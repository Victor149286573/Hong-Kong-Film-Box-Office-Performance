```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('~/Documents/yr2 term2/STAT2011/2021-2011_weeks.xlsx')
df=df.drop_duplicates(subset=['week','name'])
df.head()
```

```python
df_gross=df[['name','gross']]
```

```python
df_gross.head()
```

```python
total_gross=df_gross.groupby(df_gross['name'])['gross'].agg(['sum'])
top=total_gross.sort_values(by='sum',ascending=False)[:10]
top
```

```python
rank_gross=total_gross.sort_values('sum',ascending=False)[:100]
rank_gross['rank']=np.arange(1,101)
```

```python
rank_gross.to_csv('rank_gross.csv')
```

```python
top1=df[df['name']=='Spider-Man: No Way Home']
top2=df[df['name']=='Avengers: Age of Ultron']
top3=df[df['name']=='Iron Man 3']
top4=df[df['name']=='Eternals']
top5=df[df['name']=='Jurassic World']
top6=df[df['name']=='Transformers: Age of Extinction']
top7=df[df['name']=='The Avengers']
top8=df[df['name']=='Transformers: Dark of the Moon']
top9=df[df['name']=='Anita']
top10=df[df['name']=='Shang-Chi and the Legend of the Ten Rings']
```

```python
from pyecharts import options as opts
from pyecharts.charts import *

#Bar plot for the top10 movies' total gross
c = (
    Bar()
    .add_xaxis(list(top.index))
    .add_yaxis("Total gross", list(top['sum']))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-15)),
        title_opts=opts.TitleOpts(title="The total gross of the top10 popular movies")
    )
    .render("bar_total_gross.html")
)
```

```python
#Box plot for the top10 movies' total gross
c = Boxplot()
c.add_xaxis(["top10 total gross"])
c.add_yaxis("total gross", c.prepare_data([list(top['sum'])]))
c.set_global_opts(title_opts=opts.TitleOpts(title="Box plot for the top10 movies' total gross"))
c.render("boxplot_total_gross.html")
```

```python
#Linechart for top10 movies' total gross and week gross
c = (
    Line()
    .add_xaxis(list(top1['week'].astype(str)))
    .add_yaxis("week gross", list(top1['gross']))
    .add_yaxis("total gross", list(top1['gross'].cumsum()))
    .set_global_opts(title_opts=opts.TitleOpts(title="The 1st popular movie's gross",
                                               subtitle='Spider-Man: No Way Home'))
    .render("line_top1.html")
)
```

```python
#Piechart of the years of top10 movies
years=[]
for i in list(top.index):
    sub=df[df['name']==i]
    year=int(sub.drop_duplicates(subset='year')['year'])
    years.append(year)
years.sort()
years_count=pd.DataFrame(pd.DataFrame(years).value_counts())
```

```python
c = (
    Pie()
    .add(
        "",
        [list(z) for z in zip(['2021', '2015', '2011', '2012', '2013', '2014'],
                              list(pd.DataFrame(years).value_counts()))],
        radius=["30%", "75%"],
        center=["50%", "50%"],
        rosetype="radius",
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Piechart of years"))
    .render("pie_years.html")
)
```

```python
num_weeks=[]
for i in list(top.index):
    sub=df[df['name']==i]
    weeks=list(sub['week'])
    num=weeks[-1]-weeks[0]+1
    num_weeks.append(num)
```

```python
top['num_weeks']=num_weeks
sort_week=top.sort_values(by='num_weeks')
sort_week
```

```python
c = (
    Line()
    .add_xaxis(list(sort_week['num_weeks'].astype(str)))
    .add_yaxis("total gross", list(sort_week['sum']))
    .set_global_opts(title_opts=opts.TitleOpts(title="The relationship between release weeks and gross"))
    .render("Time-gross.html")
)
```
